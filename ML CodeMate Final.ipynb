```python
def achieve(i):
    if i =='None':
        return 0
    else:
        return 1    
```

```python
df['Achievements2']=df['Achievements:'].apply(achieve)
```

```python
df.head()
```

```python
cat=['Gender', 'Year of Study', 'Branch', 'Type of Schooling <10th',
       '% in Class 10th', 'Location of Class 10th', 'Type of Schooling 12th',
       '% in Class 12th', 'Location of Class 12th', 'Dropped 1y after School',
       'Want to be 1st generation graduate',
       'Involved in additional volunteer',
       'The number of programming languages',
       'Involved in specific technology?', 'Data structure and algorithms?',
       'Github?', 'Associated with any developer Community',
       'Developer Communities', 'Hackathons and other competitions',
       'Software based projects', 'Hardware based project', 'Pitch',
       'Any additional skills']
```

#### 1) Non achievement and achievement

```python
df['Achievements2'].value_counts().plot(kind='bar',legend=True);
```

#### INFERENCE:
    It is evident and positive observation that there are more number of students who have achieved something in           coding domain.
    Though there is not huge difference between the number of students who have achievements and others who do not.

#### 2) Histplot by Non achievement and achievement

```python
fig = plt.figure(figsize=(20,30))

for i in range(len(cat)):
    plt.subplot(8,3,i+1)
    sns.histplot(data=df,y=df[cat[i]],hue=df['Achievements2'])
    
plt.tight_layout()
plt.show()
```

```python
df['Hackathons and other competitions'].value_counts().plot(kind='bar',legend=True);
```

#### INFERENCE:
    There is a need to inculcate coding competitions among students in order to increase their skills and make them aware about where they stand when compared to people of same domain.

```python
df['Github?'].value_counts().plot(kind='bar',legend=True);
```

```python
df['Data structure and algorithms?'].value_counts().plot(kind='pie',legend=True);
```

#### INFERENCE:
    It is evident that students already know about DSA just their expertise differ.

```python
Dsa_Hackathon=df.groupby(['Hackathons and other competitions','Data structure and algorithms?']).size().unstack()
```

```python
Dsa_Hackathon.plot.bar(stacked=True, figsize=(10,10))
```

#### INFERENCE:
    It is observed that students who haven't participated in Hackathons still have intermediate knowledge of DSA, 
    they just haven't put it to practise.

```python
Git_Dev=df.groupby(['Associated with any developer Community','Github?']).size().unstack()
```

```python
Git_Dev.plot.bar(stacked=True, figsize=(10,10))
```

#### INFERENCE:
It's observed that the more students have knowledge of Github but they are still not part of any commmunity.
